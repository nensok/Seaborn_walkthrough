```python
import seaborn as sns
```

```python
# getting datasets from seaborn
sns.get_dataset_names()
```

```python
# Loading Datasets from seaborn
tips = sns.load_dataset('tips')
planets = sns.load_dataset('planets')
taxis = sns.load_dataset('taxis')
titanic = sns.load_dataset('titanic')
```

```python
# Scatter plot using seaborn
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='time', size='size', palette='YlGnBu')
```

```python
# Histogram plot in Seaborn
sns.histplot(tips['tip'], kde=True, bins=30)
```

```python
sns.displot(tips['tip'], kde=True, bins=30)
```

```python
# Barplots in seaborn
sns.barplot(x='sex', y='tip', data=tips)
```

```python
#Box plot in seaborn
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
# Stripplot in seaborn
sns.stripplot(x='day', y='tip', data=tips, hue='sex', dodge=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', shade=True)
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```

```python
titanic.corr()
```

```python
# Plotting a heatmap in seaborn
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu')
```

```python
sns.clustermap(titanic.corr())
```
